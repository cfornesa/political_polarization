```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# from google.colab import drive
# drive.mount('/content/drive')
```

```python
# Import the datasets
# results_path = '/content/drive/My Drive/Online MSDS/MOD C2/Political Polarization/results/'
results_path = 'results/'

linear_regression = pd.read_csv(f"{results_path}model1_linear-model-summary.csv")
feature_regression = pd.read_csv(f"{results_path}model2_feature-regularization-summary.csv")
stepwise_regression = pd.read_csv(f"{results_path}model3_stepwise-regression-summary.csv")
logistic_regression = pd.read_csv(f"{results_path}model4_logistic-regression-summary.csv")
support_vector_regression = pd.read_csv(f"{results_path}model5_svm-regression-summary.csv")
support_vector_classification = pd.read_csv(f"{results_path}model5_svm-classification-summary.csv")
decision_tree_regression = pd.read_csv(f"{results_path}model6_dt-regression-summary.csv")
decision_tree_classification = pd.read_csv(f"{results_path}model6_dt-classification-summary.csv")
random_forest_regression = pd.read_csv(f"{results_path}model6_rf-regression-summary.csv")
random_forest_classification = pd.read_csv(f"{results_path}model6_rf-classification-summary.csv")
knn_regression = pd.read_csv(f"{results_path}model8_knn-regression-summary.csv")
knn_classification = pd.read_csv(f"{results_path}model8_knn-classification-summary.csv")
gb_regression = pd.read_csv(f"{results_path}model9_gb-regression-summary.csv")
gb_classification = pd.read_csv(f"{results_path}model9_gb-classification-summary.csv")
kmeans_clustering = pd.read_csv(f"{results_path}model10_kmeans-clustering-summary.csv")
```

## Standardize the data

```python
# Change "Regularizer" to "Regularization"
logistic_regression.rename(columns = {'Regularizer':'Regularization', 'Solver': 'Kernel'}, inplace = True)

# Change all "DGHS" values to "Dynamically Generated Hate Speech"
logistic_regression["Model"] = "Logistic Regression"
logistic_regression.loc[logistic_regression['Dataset'] == 'DGHS', 'Dataset'] = 'Dynamically Generated Hate Speech'
logistic_regression.loc[logistic_regression['Dataset'] == 'US2020HS', 'Dataset'] = 'US Elections 2020 Election Hate Speech'
logistic_regression = logistic_regression[["Model", "Dataset", "Accuracy", "F1-score", "Ratio"]]

support_vector_classification["Model"] = "Support Vector Machine"
support_vector_classification["Ratio"] = 0.0
support_vector_classification.rename(columns = {'F1 Score':'F1-score'}, inplace = True)

# Switch the position of "Kernel" and "Regularization"
support_vector_classification = support_vector_classification[["Model", "Dataset", "Accuracy", "F1-score", "Ratio"]]

support_vector_regression["Model"] = "Support Vector Machine"
support_vector_regression = support_vector_regression.drop(columns=['Kernel', 'Regularization'])
support_vector_regression = support_vector_regression[["Model", "Dataset", "RMSE", "R2"]]

# Remove the "Best Parameters" column from decision_tree_regression and decision_tree_classification
decision_tree_regression = decision_tree_regression.drop(columns=['Best Parameters'])
decision_tree_classification = decision_tree_classification.drop(columns=['Best Parameters', 'Kernel', 'Regularization'])

# Remove the "Best Parameters" column from random_forest_regression and random_forest_classification
random_forest_regression = random_forest_regression.drop(columns=['Best Parameters'])
random_forest_classification = random_forest_classification.drop(columns=['Best Parameters', 'Kernel', 'Regularization'])

# Remove the "Best Parameters" column from knn_regression and knn_classification
knn_regression = knn_regression.drop(columns=['Best Parameters'])
knn_classification = knn_classification.drop(columns=['Best Parameters', 'Kernel', 'Regularization'])

# Remove the "Best Parameters" column from gb_regression and gb_classification
gb_regression = gb_regression.drop(columns=['Best Parameters'])
gb_classification = gb_classification.drop(columns=['Best Parameters', 'Kernel', 'Regularization'])

# Remove the "Best Parameters" column from kmeans_clustering
kmeans_clustering = kmeans_clustering.drop(columns=['Best Parameters'])
```
