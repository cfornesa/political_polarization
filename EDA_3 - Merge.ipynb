```python
convabuse = pd.read_csv(f"{path}convabuse_final.csv")
dynamically_generated_hate_speech = pd.read_csv(f"{path}dynamically_generated_hate_speech_dataset_final.csv")
online_abusive_attacks = pd.read_csv(f"{path}online_abusive_attacks_final.csv")
us_elections_2020_hate_speech = pd.read_csv(f"{path}us_elections_2020_hate_speech_final.csv")
mlma_hate_speech = pd.read_csv(f"{path}MLMA_hate_speech_final.csv")
```

```python
# Merge Convabuse and MLMA Hate Speech on "abuse_level" using inner join
convabuse_mlma = pd.concat([convabuse, mlma_hate_speech], ignore_index=True, sort=False)
```

```python
convabuse_mlma.head()
```

```python
# Drop the directness and columns with the term "sentiment"
convabuse_mlma = convabuse_mlma.drop(columns=[col for col in convabuse_mlma.columns if 'directness' in col or 'sentiment' in col])
```

```python
# Export the merged dataset to a CSV file
convabuse_mlma.to_csv(f"{path}convabuse_mlma_merged.csv", index=False)
```
